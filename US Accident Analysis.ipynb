```python
#### START TIME
df.Start_Time[0]
## ^^ This output is currently in string, we want this to convert into date and time. 
df.Start_Time = pd.to_datetime(df.Start_Time)
df.Start_Time[0]
```

```python
sns.distplot(df.Start_Time.dt.hour, bins = 24, kde = False, norm_hist = True) #BINS TO MAX OUT AT 24hrs
```

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins = 7, kde = False, norm_hist = True) #BINS TO MAX OUT AT 7 days
```

```python
###PATTERN FOR SUNDAY - THE PEAK IS BETWEEN 1pm - 6pm. 
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour, bins = 24, kde = False, norm_hist = True)
```

```python
###PATTERN FOR SATURDAY IS SIMILAR TO SUNDAY - PEAK IS BETWEEN 12pm to 6pm. 
saturdays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 5]
sns.distplot(saturdays_start_time.dt.hour, bins = 24, kde = False, norm_hist = True)
```

```python
### PATTERN ON MONDAY IS SIMILAR TO PATTERN FOR ALL WEEKDAYS - PEAK IS BETWEEN 3pm - 5pm. 
mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 1]
sns.distplot(mondays_start_time.dt.hour, bins = 24, kde = False, norm_hist = True)
```

```python
#### THE PATTERN SHOWS THAT THE MONDAYS ARE NOT AS BUSY DURING JAN - AUGUST. Things start getting busy after. 
###The data does not have Jan 2016 data at this time, it only covers data from Feb 2016. 
sns.distplot(df.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```

```python
''' in order to investigate if the previous graph is an accurate representation of the traffic through Jan - Dec '''
###OUR PATTERN LOOKS QUITE SIMILAR OVERALL. BETWEEN 2018, 2019 and the previous graph. 
df_2018 = df[df.Start_Time.dt.year == 2019]
sns.distplot(df_2018.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```
